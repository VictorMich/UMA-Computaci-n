# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([5,2,3])
plt.show()
```

```python
z = np.arange(1,11)
c = -(np.arange(1,11))**2
plt.scatter(z,c, color="crimson",s=500)#Puntos
plt.plot(z,c, color="k",linewidth=3)
plt.grid(alpha = 6)#Cuadrícula que tan tenue se ve
plt.title("Mi increíble gráfica")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.show()
```

```python
x = np.arange(0,6,0.10)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(-1,1,200)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
x = (np.linspace(-2,2))
for i in range(1,7):
    plt.plot(x,x**i)
    
```

```python
x = np.linspace(1e-6,1)
y = -x*np.log(x)
plt.plot(x,y)
```
